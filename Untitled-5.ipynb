***fonction stemming***

```python
def stemming(content):
  stemmed_content = re.sub('[^a-zA-Z]',' ',content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content
```

utilisation de random_forest

```python
from sklearn.ensemble import RandomForestClassifier

# Modèle Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)

# Prédictions
y_pred_rf = rf_model.predict(x_test)


```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
cm = confusion_matrix(y_test,y_pred_rf )
cm
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rf_model.classes_)
disp.plot(cmap="Blues")
disp.ax_.set_title("Matrice de confusion")
plt.show()

# Rapport de classification (précision, rappel, F1-score)
print("Rapport de classification :")
print(classification_report(y_test, y_pred_rf ))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Définir le modèle de base
logistic_model = LogisticRegression()

# Définir la grille des hyperparamètres à tester
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', None],  # Type de régularisation
    'C': [0.01, 0.1, 1, 10, 100],                # Force de la régularisation
    'solver': ['lbfgs', 'liblinear', 'saga'],    # Solvers disponibles
    'max_iter': [100, 500, 1000]                 # Nombre maximum d'itérations
}

# Définir GridSearchCV
grid_cv = GridSearchCV(estimator=logistic_model, param_grid=param_grid, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)

# Ajuster le modèle aux données d'entraînement
grid_cv.fit(x_train, y_train)


# Meilleurs hyperparamètres trouvés
print("Meilleurs hyperparamètres :", grid_cv.best_params_)

# Évaluer sur les données de test
best_model = grid_cv.best_estimator_
y_pred = best_model.predict(x_test)

print("Accuracy sur les données de test :", accuracy_score(y_test, y_pred))
```

```python
from googleapiclient.discovery import build
import pandas as pd


# Remplacez par votre clé API
api_key = "AIzaSyDuuKa4rOXUUAWp1V-wA2JpAV4mIr6obUs"

# Initialisation du client YouTube
youtube = build("youtube", "v3", developerKey=api_key)

# Fonction pour récupérer les informations de la vidéo
def obtenir_infos_video(video_id):
    response = youtube.videos().list(
        part="snippet,statistics",
        id=video_id
    ).execute()

    video_data = response['items'][0]
    titre = video_data['snippet']['title']
    description = video_data['snippet']['description']
    nb_vues = video_data['statistics']['viewCount']
    nb_likes = video_data['statistics'].get('likeCount', 'N/A')  # Gestion des valeurs manquantes
    nb_commentaires = video_data['statistics'].get('commentCount', 'N/A')
    return {
        "Titre": titre,
        "Description": description,
        "Nombre de vues": nb_vues,
        "Nombre de likes": nb_likes,
        "Nombre de commentaires": nb_commentaires
    }

# Fonction pour récupérer les commentaires avec pagination
def obtenir_commentaires(video_id, max_results=100):
    commentaires = []
    next_page_token = None
    page_count = 0

    while True:
        page_count += 1
        print(f"Récupération de la page {page_count}...")  # Pour suivre le progrès

        # Appel à l'API YouTube pour récupérer une page de commentaires
        response = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=max_results,
            pageToken=next_page_token
        ).execute()

        # Extraire les commentaires de la réponse
        for item in response['items']:
            commentaire = item['snippet']['topLevelComment']['snippet']['textDisplay']
            auteur = item['snippet']['topLevelComment']['snippet']['authorDisplayName']
            commentaires.append({"Auteur": auteur, "Commentaire": commentaire})

        # Vérifier si une page suivante existe
        next_page_token = response.get('nextPageToken')
        if not next_page_token:  # Si plus de pages, on arrête la boucle
            print(f"Tous les commentaires récupérés : {len(commentaires)}")
            break

    return commentaires
 
# ID de la vidéo
video_id = "0FtcHjI5lmw"  # Remplacez par un ID réel
video_id = input("Entrez l'ID de la vidéo : ")[-11:]
# Obtenir les informations de la vidéo
info_video = obtenir_infos_video(video_id)

# Obtenir les commentaires
commentaires = obtenir_commentaires(video_id)
titre_video = info_video["Titre"]

# Créer des tableaux avec pandas
df_info = pd.DataFrame([info_video])  # Tableau avec une seule ligne pour les infos vidéo
df_commentaires = pd.DataFrame(commentaires)  # Tableau des commentaires
# Afficher les tableaux
print("Informations sur la vidéo :")
print(df_info)
print("\nCommentaires :")
print(f"Nombre total de commentaires récupérés : {len(df_commentaires)}")
print(df_commentaires)  # Afficher tous les commentaires récupérés
new_comments = df_commentaires['Commentaire'].to_list()
new_comments_processed = [stemming(comment) for comment in new_comments]
new_comments_vectors = loaded_vectorizer.transform(new_comments_processed)
predictions = loaded_model.predict(new_comments_vectors)
print(predictions)
import matplotlib.pyplot as plt

# Compter les prédictions positives et négatives
positive_count = sum(predictions == 1)  # Suppose que 1 = positif
negative_count = sum(predictions == 0)  # Suppose que 0 = négatif

# Calculer les pourcentages
total = len(predictions)
positive_percentage = (positive_count / total) * 100
negative_percentage = (negative_count / total) * 100

# Créer un diagramme circulaire
labels = ['Positifs', 'Négatifs']
sizes = [positive_percentage, negative_percentage]
colors = ['#4CAF50', '#F44336']  # Vert pour positif, rouge pour négatif
explode = (0.1, 0)  # Mettre en avant le segment positif

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title(f"Analyse des Sentiments pour la Vidéo (: {titre_video})")
plt.axis('equal')  # Assure que le cercle est parfaitement rond
plt.show()


```
